```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv("insurance_data.csv")

# This Dataset have only 1 Feature(Age)
print("First 5 rows of the DataFrame:")
print(df.head())
```

```python
plt.scatter(df.age,df.bought_insurance,marker='+',color='red')
plt.show()
```

```python
# Splitting Dataset for testing and training
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df[['age']],df.bought_insurance,train_size=0.8)
```

```python
# Training Model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

model.fit(X_train, y_train)
```

```python
model.predict([[40]])

# 0 = NO | 1 = YES
# we can try different values like 35 or 50
```

```python
# --- INTERNAL WORKING IN NEURON ---
# y = m * x + b
m = model.coef_[0]
b = model.intercept_

# Print the values
print(f"m (coefficient) = {m}")
print(f"b (intercept) = {b}")

# The linear regression equation is y = m * x + b
print(f"\nThe equation is: y = {m} * x + {b}")
```

```python
import math
def sigmoid(y):
  return 1 / (1 + math.exp(-y))

def prediction_function(age):
    # y = m * x + b
    # 0.04150133 ~ 0.042 and -1.52726963 ~ -1.53
    z = 0.042 * age - 1.53
    y = sigmoid(z)
    return y
```

```python
prediction_function(40)
```
