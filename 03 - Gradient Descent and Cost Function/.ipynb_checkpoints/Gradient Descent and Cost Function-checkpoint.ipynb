## Cost Function (MSE)

Mean Squared Error (MSE) is used to measure error.

$$J(m, b) = \frac{1}{n} \sum (y_i - (mx_i + b))^2$$

```python
def gradient_descent(x,y):
    # Eqn of line y = mx + b
    # current slope and current constant
    m_curr = b_curr = m_best = b_best = 0
    n = len(x)
    iterations = 10000
    learning_rate = 0.08
    lowest_cost = 10000
    
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
    
        # Calculating Cost
        cost = (1/n) * sum([val**2 for val in (y - y_predicted)])

        if(cost < lowest_cost):
            m_best = m_curr
            b_best = b_curr

        # we know len(x) and len(y) is same and dataset is complete
        # lets calculate m derivative 
        md = -(2/n) * sum(x * (y - y_predicted))
        # lets calculate m derivative
        bd = -(2/n) * sum(y - y_predicted)
    
        # lets move to next pair of m and b
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
    
        # Printing current pair of m and b
        # print("m {}, b {}, cost {}, iteration{}".format(m_curr, b_curr, cost, i))

    
    print("m {}, b {}".format(m_best, b_best))
```

```python
gradient_descent(x,y)

#Note: Correct Answer for this problem is m = 2 and b = 3
```
