```python
# Sigmoid Function
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
# Trying Sigmoid Function (Best for output layer)
print(sigmoid(100))
print(sigmoid(5))
print(sigmoid(-5))
print(sigmoid(0.5))

# Always gives output between [0,1]
```

```python
# Tanh Function
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
# Trying Tanh Function (Best for hidden layer)
print(tanh(100))
print(tanh(5))
print(tanh(-5))
print(tanh(0.5))

# Always gives output between [-1,1]
```

```python
# ReLU Function
def relu(x):
    return max(0, x)
```

```python
# Trying ReLU Function (Best when you don't know what to apply)
print(relu(100))
print(relu(5))
print(relu(-5))
print(relu(0.5))

# Always gives output between [0,∞)
```

```python
# leaky ReLU Function
def leaky_relu(x):
    return max(0.1*x, x)
```

```python
# Trying ReLU Function
print(leaky_relu(100))
print(leaky_relu(5))
print(leaky_relu(-5))
print(leaky_relu(0.5))

# Always gives output between (∞,∞)
```
